```python
import pandas as pd
data_set = pd.DataFrame()
for i in range(19):
    df = pd.read_csv(filepath_or_buffer='./all_2018/'+str(i)+'.csv',encoding='gbk')
    df = df[df['year']==2018]
    df = df[(df['sex_id']!=3)&(df['age_id']!=22)]
    data_set = pd.concat([data_set,df])
data_set
    
```

```python
data_set = data_set.drop(['cause_id','cause_name'],axis=1)
death_18 = data_set[data_set['measure_id']==1]
death_18.to_csv('death_18.csv',encoding='gbk',index=False)
DALYs_18 = data_set[data_set['measure_id']==2]
DALYs_18.to_csv('DALYs_18.csv',encoding='gbk',index=False)
YLDs_18 = data_set[data_set['measure_id']==3]
YLDs_18.to_csv('YLDs_18.csv',encoding='gbk',index=False)
YLLs_18 = data_set[data_set['measure_id']==4]
YLLs_18.to_csv('YLLs_18.csv',encoding='gbk',index=False)
```
