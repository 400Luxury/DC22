```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("joinpoint/China_DM_DALYs.data.csv",encoding='gbk')
df
```

```python
def switch_age(age):
    if (age == '75'):
        return '75+'
    else:
        return age+'-'+str(int(age)+4)
        
def drawgraph(sex, age, linecolor):
    df_c = df.loc[(df['sex'].str[0]==sex) & (df['age'].str[0:2]==age)]
    plt.plot(df_c['year'],df_c['Model'],c=linecolor,label=switch_age(age))
    plt.xlabel('year')
    plt.ylabel('DALYs')
```

```python
plt.figure(figsize=(18,13))
plt.subplot(2,2,1)
drawgraph('F','25','#FFBBFF')
drawgraph('F','30','#EEA9B8')
drawgraph('F','35','#FF0000')
drawgraph('F','40','#FF34B3')
drawgraph('F','45','#8B636C')
plt.legend()
plt.title('Female.25-50 years old')
plt.subplot(2,2,3)
drawgraph('F','50','#FFBBFF')
drawgraph('F','55','#EEA9B8')
drawgraph('F','60','#FF0000')
drawgraph('F','65','#FF00FF')
drawgraph('F','70','#FF34B3')
drawgraph('F','75','#8B636C')
plt.legend()
plt.title('Female.50+ years old')
plt.subplot(2,2,2)
drawgraph('M','25','#E0EEE0')
drawgraph('M','30','#B0E2FF')
drawgraph('M','35','#8968CD')
drawgraph('M','40','#00B2EE')
drawgraph('M','45','#00688B')
plt.legend()
plt.title('Male.25-50 years old')
plt.subplot(2,2,4)
drawgraph('M','50','#E0EEE0')
drawgraph('M','55','#B0E2FF')
drawgraph('M','60','#8968CD')
drawgraph('M','65','#00B2EE')
drawgraph('M','70','#00688B')
drawgraph('M','75','#90EE90')
plt.legend()
plt.title('Male.50+ years old')
plt.show()
```

```python
df_s = pd.read_csv("joinpoint/export.data.csv",encoding='gbk')
df_f = df_s[df_s['sex']=='Female']
df_m = df_s[df_s['sex']=='Male']
plt.figure(figsize=(9,6.5))
plt.plot(df_f['year'],df_f['Model'],c='pink',label='Female')
plt.plot(df_m['year'],df_m['Model'],c='grey',label='Male')
plt.scatter(df_f['year'],df_f['val'],c='red')
plt.scatter(df_m['year'],df_m['val'],c='blue')
plt.xlabel('year')
plt.ylabel('DALYs')
plt.title('T2D: Female vs Male')
plt.legend()
plt.show()
```
